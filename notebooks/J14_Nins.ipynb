## Imports

```python
from copy import deepcopy
```

```python
day = "14"
```

```python
with open(f"input_day{day}.txt") as f:
    robots = f.readlines()
    robots = list(map(lambda x: x.replace("\n", "").split(), robots))

len(robots)
```

```python
positions = [[int(x) for x in robot[0][2:].split(",")] for robot in robots]
velocities = [[int(x) for x in robot[1][2:].split(",")] for robot in robots]
```

## Part 1

```python
def predict_position_after_x_s(
    positions: list,
    velocities: list,
    width: int= 11,
    height: int = 7,
    nb_seconds: int = 100,
):
    new_positions = deepcopy(positions)
    for i, pos in enumerate(positions):
        new_positions[i][0] = (pos[0] + nb_seconds * velocities[i][0]) % width
        new_positions[i][1] = (pos[1] + nb_seconds * velocities[i][1]) % height
    
    return new_positions
```

```python
new_positions = predict_position_after_x_s(positions, velocities, width=101, height=103)
```

```python
def get_safesty_factor(
    positions: list,
    width: int= 11,
    height: int = 7,
):
    area_1, area_2, area_3, area_4 = 0, 0, 0, 0
    middle_width = (width-1)//2
    middle_height = (height-1)//2
    for pos in positions:
        if pos[0] < middle_width:
            if pos[1] < middle_height:
                area_1 += 1
            elif pos[1] > middle_height:
                area_3 += 1
        elif pos[0] > middle_width:
            if pos[1] < middle_height:
                area_2 += 1
            elif pos[1] > middle_height:
                area_4 += 1

    return area_1 * area_2 * area_3 * area_4
```

```python
get_safesty_factor(new_positions, width=101, height=103)
```

## Part 2

```python
def print_robots(positions: list, width: int = 101, height: int = 103):
    grid = [[0 for j in range(height)] for i in range(width)]
    for pos in positions:
        grid[pos[0]][pos[1]] += 1
    
    # for row in grid:
    #     print(row)
    return grid
```

```python
def find_tree(
    positions: list,
    velocities: list,
    width: int= 101,
    height: int = 103,
):
    grid = [[]]
    consecutive = 0
    nb_seconds = 0
    while consecutive < width//5:  # Proxy to check if there is a tree
        nb_seconds += 1
        grid = print_robots(
            predict_position_after_x_s(positions, velocities, width=width, height=height, nb_seconds=nb_seconds)
        )

        consecutive = 0
        for row in grid:
            max_row = 0
            for el in row:
                if el > 0:
                    max_row += 1
                else:
                    consecutive = max(max_row, consecutive)
                    max_row = 0

        if nb_seconds % 1000 == 0:
            print(nb_seconds)
        if nb_seconds > 10000:
            break
    for row in grid:
        print(row)
    return nb_seconds
```

```python
find_tree(
    positions,
    velocities,
    width = 101,
    height = 103
)
```
